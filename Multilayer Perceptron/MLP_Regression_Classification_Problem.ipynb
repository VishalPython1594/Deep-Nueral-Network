```python
# Importing the basis libraries
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('/content/HousingData.csv')
dataset.head()
```

```python
dataset.isnull().sum()/len(dataset)*100
```

```python
dataset.info()
```

```python
# applying knn imputation method
from sklearn.impute import KNNImputer, SimpleImputer
# SimpleImputer is for univariate imputation (mean, median and mode)
# KNNImputer is working basis the K nearest neighbors approach - more famous
imputer = KNNImputer(n_neighbors=5)
x_impute = pd.DataFrame(imputer.fit_transform(dataset), columns=dataset.columns)
x_impute
```

```python
x_impute.isnull().sum()
```

```python
# splitting the data into x and y
x = x_impute.iloc[:,0:-1]
y = x_impute.iloc[:,-1]
```

```python
x.head()
```

```python
x.describe()
```

```python
x.isnull().sum()
```

```python
x.shape
```

```python
y.head()
```

# Feature Scaling is mandatory steps in Deep Neural Network
# Please don't process your model without feature scaling otherwise you will get pathetic result

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x_scaler
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Using MultiLayer Perceptron Model (ANN or DNN)

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense # Fully connected layers
```

```python
dnn = Sequential()
# Hidden layer 1 connected with input dataset
dnn.add(Dense(28, activation='relu', input_dim=x_train.shape[1]))
# vanishing gradient and exploding gradient problem
# Hidden Layer 2
dnn.add(Dense(28, activation='relu'))
# Output
dnn.add(Dense(1))
# check parameter
dnn.summary()
```

```python
dnn.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mae'])
history = dnn.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```

# Encoding with categorical variable
# Gender and Geography

```python
mydata['Gender'] = mydata['Gender'].astype('category')
mydata['Gender'] = mydata['Gender'].cat.codes
mydata.head()
```

```python
mydata.head()
```

```python
mydata = pd.get_dummies(mydata, columns=['Geography'], drop_first= True)
```

```python
"""
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])], remainder='passthrough')
x = pd.DataFrame(ct.fit_transform(mydata))
x
"""
```

```python
mydata.head()
```

```python
mydata['Geography_Germany'] = np.where(mydata['Geography_Germany']==False, 0, 1)
mydata['Geography_Spain'] = np.where(mydata['Geography_Spain']==False, 0, 1)
```

```python
mydata.head()
```

```python
# split the data into x and y
x = mydata.drop('Exited', axis=1)
y = mydata[['Exited']]
```

```python
x.head()
```

```python
y.head()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
x_scaler
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Deep Neural Network/MultiLayer Perceptron

```python
dnn = Sequential()
dnn.add(Dense(32, activation='relu', input_dim=11))
dnn.add(Dense(16, activation='relu'))
dnn.add(Dense(1, activation='sigmoid'))
dnn.summary()
```

```python
dnn.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
history = dnn.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```
